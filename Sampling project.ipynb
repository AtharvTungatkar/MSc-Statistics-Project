```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nifty50=pd.read_csv('ind_nifty50list.csv')
```

```python
nifty50['Industry'].value_counts()
```

```python
sectors=['Financial Services','Automobile and Auto Components','Healthcare','Fast Moving Consumer Goods','Information Technology']
```

```python
financial_services=nifty50[nifty50['Industry']==sectors[0]].reset_index()
automobile_and_auto_components=nifty50[nifty50['Industry']==sectors[1]].reset_index()
healthcare=nifty50[nifty50['Industry']==sectors[2]].reset_index()
fast_moving_consumer_goods=nifty50[nifty50['Industry']==sectors[3]].reset_index()
information_technology=nifty50[nifty50['Industry']==sectors[4]].reset_index()
```

```python
lengths=[len(financial_services),len(automobile_and_auto_components),len(healthcare),len(fast_moving_consumer_goods),len(information_technology)]
```

```python
proportion=[]
sample_size=15
sample=[]
for i in range(len(sectors)):
    sample.append(sample_size*(lengths[i]/sum(lengths)))
    proportion.append((lengths[i]/sum(lengths)))
```

```python
sample
```

```python
sample_final=[5,3,2,2,3]
```

```python
plt.figure(figsize=(20,10))
plt.bar(x=sectors,height=proportion)
plt.grid()
plt.title('Sectorwise Population Proportion')
plt.ylabel('Proportion of stocks')
```

```python
plt.figure(figsize=(20,10))
plt.bar(x=sectors,height=sample_final)
plt.grid()
plt.title('Sectorwise Sample Allocation')
plt.ylabel('Number of Stocks')
```

```python
# Create the figure and axes
fig, axes = plt.subplots(1, 2, figsize=(15, 10))

# Plot the first pie chart
axes[0].pie(x=proportion, labels=sectors, autopct='%1.1f%%')
axes[0].set_title('Sectorwise Population Proportions')

# Plot the second pie chart
axes[1].pie(x=sample_final, labels=sectors, autopct='%1.1f%%')
axes[1].set_title('Sectorwise Sample Proportion')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Sectorwise Sample

financial_services_sample=financial_services.iloc[list(np.random.choice(np.arange(1, len(financial_services)), size=5, replace=False))]

automobile_and_auto_components_sample=automobile_and_auto_components.iloc[list(np.random.choice(np.arange(1, len(automobile_and_auto_components)), size=sample_final[1], replace=False))]

healthcare_sample=healthcare.iloc[list(np.random.choice(np.arange(1, len(healthcare)), size=sample_final[2], replace=False))]

fast_moving_consumer_goods_sample=fast_moving_consumer_goods.iloc[list(np.random.choice(np.arange(1, len(fast_moving_consumer_goods)), size=sample_final[3], replace=False))]

information_technology_sample=information_technology.iloc[list(np.random.choice(np.arange(1, len(information_technology)), size=sample_final[4], replace=False))]
```

```python
final_sample_data=pd.concat([financial_services_sample,automobile_and_auto_components_sample,healthcare_sample,
                            fast_moving_consumer_goods_sample,information_technology_sample],axis=0).reset_index().drop(['index','level_0'],axis=1)
```

```python
final_sample_data['tickers']=final_sample_data['Symbol']+'.NS'
```
